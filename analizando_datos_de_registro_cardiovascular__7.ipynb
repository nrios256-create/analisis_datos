```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
regicor = pd.read_csv('regicor.csv') #carga de datos
regicor.info()
regicor.shape
```

```python
regicor.head()
```

```python
columnas_numericas =regicor.select_dtypes([float, int]).columns
print('columnas numericas')
print(columnas_numericas)
```

```python
columnas_categoricas =regicor.select_dtypes('object').columns
print('columnas categoricas')
print(columnas_categoricas)
```

```python
columnas_categoricas
```

```python
#para todas las columnas tipo objeto las convierto en category
regicor[columnas_categoricas]=regicor[columnas_categoricas].astype('category')
```

```python
#con este metodo verificamos que si paso a categoricas
regicor.info()
```

```python
#ESTADISTICA DESCRIPTIVAS

print('\n'+ '='*60)
print('estadisticas descriptivas')
print('='*60)

columnas_numericas
regicor[columnas_numericas].describe().round(2)

print('\nDistribucion por sexo:')
print(regicor['sex'].value_counts())
print(f'\nPorcentaje:')
print((regicor['sex'].value_counts(normalize=True)*100).round(2))

print('\nDistribucion por año:')
print(regicor['year'].value_counts().sort_index())

print('\nDistribucion por estado de fumador:')
print(regicor['smoker'].value_counts())
```

```python
regicor.isnull() #nos muestra los valores faltantes y nulos, true que si hay valores faltantes o nulos y viceversa
```

```python
regicor.isnull().sum() #sum me suma los true por cada una
```

```python
#VALORES FALTANTES

print('\n'+ '='*60)
print('ANALISIS DE VALORES FALTANTES')
print('='*60)

missing =regicor.isnull().sum()
missing_pct = (missing/len(regicor)*100).round(2)
missing_regicor = pd.DataFrame({'Faltantes':missing, 'Porcentaje':missing_pct}) #dataframe con valores faltantes
missing_regicor = missing_regicor[missing_regicor['Faltantes']>0].sort_values('Porcentaje', ascending=False)
print('\nVariables con valores faltantes:')
missing_regicor
```

```python
columnas_numericas
```

```python
#ANALISIS POR GRUPOS

print('\n'+ '='*60)
print('ANALISIS POR GRUPOS')
print('='*60)

print('\nEstadisticas por sexo:')
print(regicor.groupby('sex')[columnas_numericas].mean().round(2))

print('\nEstadisticas por año:')
print(regicor.groupby('year')[columnas_numericas[2:6]].mean().round(2))

print('\nEstadisticas por estado de fumador:')
print(regicor.groupby('smoker')[columnas_numericas].mean().round(2))
```

```python
#Tabulacion cruzada, la cual se realiza con variable categoricas, se pueden considerar los marginales
pd.crosstab(regicor['sex'], regicor['smoker'], margins=True)
```

```python
#Mapa de calor para la tabla cruzada

sns.heatmap(pd.crosstab(regicor['smoker'], regicor['sex']), annot=True, cmap ='flare', fmt='d')
plt.show()
```

```python
#una forma alternativa es hacer un grafico categorico
sns.catplot(data= regicor.query('year==2005'), x='sex', kind='count', col='smoker', hue='sex', palette='flare')
plt.show()
```

```python
sns.catplot(data= regicor.query('year==1995'), x='sex', y= 'chol', kind='boxen', col='histchol', hue='sex', palette='flare')
plt.show()
```
